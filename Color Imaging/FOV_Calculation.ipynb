```python
import numpy as np
import matplotlib.pyplot as plt

def field_of_view(sensor_width, focal_length):
    """Compute horizontal FOV in degrees"""
    return 2 * np.degrees(np.arctan(sensor_width / (2 * focal_length)))

def vertical_fov(sensor_height, focal_length):
    return 2 * np.degrees(np.arctan(sensor_height / (2 * focal_length)))

# Example: Full-frame sensor (36 mm width)
sensor_width = 36
sensor_height = 24
focal_lengths = [18, 35, 50, 100, 200]

fovs = [field_of_view(sensor_width, f) for f in focal_lengths]
vertical_fovs = [vertical_fov(sensor_height, f) for f in focal_lengths]


plt.figure(figsize=(8, 5))
plt.plot(focal_lengths, fovs, marker='o', label='Horizontal FOV')
plt.plot(focal_lengths, vertical_fovs, marker='x', label='Vertical FOV')
plt.title("Focal Length vs Field of View")
plt.xlabel("Focal Length (mm)")
plt.ylabel("FOV (degrees)")
plt.gca().invert_xaxis()
plt.legend()
plt.grid(True)
plt.show()
```
