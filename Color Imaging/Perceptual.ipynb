```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.color import rgb2lab

# Define perceptually similar RGB colors
# Base color: Bright red
rgb1 = np.array([255, 50, 50], dtype=np.uint8)
# Similar color: Adjusted RGB for similar LAB
rgb2 = np.array([245, 55, 45], dtype=np.uint8)  # Adjusted for smaller ΔE*

# Convert to LAB and verify similarity
lab1 = rgb2lab(rgb1.reshape(1, 1, 3))[0, 0]
lab2 = rgb2lab(rgb2.reshape(1, 1, 3))[0, 0]
diff = np.linalg.norm(lab1 - lab2)
print("LAB for Image 1:", lab1)
print("LAB for Image 2:", lab2)
print("LAB difference (ΔE*):", diff)

# Create synthetic images
image1 = np.tile(rgb1, (100, 100, 1))
image2 = np.tile(rgb2, (100, 100, 1))

# Display images
fig, axes = plt.subplots(1, 2, figsize=(8, 4))
axes[0].imshow(image1)
axes[0].set_title(f"RGB: {rgb1}\nLAB: [{lab1[0]:.2f}, {lab1[1]:.2f}, {lab1[2]:.2f}]")
axes[0].axis('off')
axes[1].imshow(image2)
print("\n")
axes[1].set_title(f"RGB: {rgb2}\nLAB: [{lab2[0]:.2f}, {lab2[1]:.2f}, {lab2[2]:.2f}]\nΔE*: {diff:.2f}")
axes[1].axis('off')
plt.suptitle("Perceptually Similar Colors in LAB Space\n", y=1.02)
plt.show()
```
